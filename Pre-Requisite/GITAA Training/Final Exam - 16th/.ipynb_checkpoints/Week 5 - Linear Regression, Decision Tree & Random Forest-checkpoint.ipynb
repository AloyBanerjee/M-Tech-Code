# NYC Data

```python
#Common library
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import math

#Linear regression specific library 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error



#Load the data from csv file 

df_NYC = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\nyc.csv')
print(df_NYC.head())
#print(df_NYC.tail())
#print(df_NYC.describe())
#print(df_NYC.info())

#Linear Regression for whole dataset
input_NYC_param = df_NYC[['Food','Decor','Service','East']]
output_NYC_param = df_NYC[['Price']]
model_NYC_Full = LinearRegression()
model_NYC_Full.fit(input_NYC_param,output_NYC_param)

#Prediction based on build model
predicted_output_NYC_Model = model_NYC_Full.predict(input_NYC_param)
#print(predicted_output_NYC_Model)

mean_output_NYC_Model = np.repeat(np.mean(df_NYC['Price']), len(df_NYC['Price']))
#print(mean_output_NYC_Model)

#Validation & Performance of the model
print('Coefficient', model_NYC_Full.coef_, sep=':')
print('Intercept', model_NYC_Full.intercept_, sep=':')
print('Score (RSquared)', model_NYC_Full.score(input_NYC_param,output_NYC_param),sep=':')

print('RMSE', np.sqrt(mean_squared_error(output_NYC_param, predicted_output_NYC_Model)))
print('Base RMSE', np.sqrt(mean_squared_error(output_NYC_param, mean_output_NYC_Model)))

print(df_NYC.corr())

print('---------------------End of full model linear regression----------------------')
#Linear Regression for only Price and Food 

input_NYC_OnlyFood = df_NYC[['Food']]
output_NYC_param_Onlyfood = df_NYC[['Price']]
model_NYC_Food = LinearRegression()
model_NYC_Food.fit(input_NYC_OnlyFood, output_NYC_param_Onlyfood)

#Prediction of the output price based on the newly created model
predicted_output_NYC_OnlyFood_Model = model_NYC_Food.predict(input_NYC_OnlyFood)
#print(predicted_output_NYC_OnlyFood_Model)

output_mean_NYC_param_OnlyFood = np.repeat(np.mean(df_NYC['Price']), len(df_NYC['Price']))
#print(output_mean_NYC_param_OnlyFood)

#Validation & Performance based on the newly created model with only food
print('Coefficient', model_NYC_Food.coef_, sep=':')
print('Intercept', model_NYC_Food.intercept_, sep=':')
print('Score (RSquared)(Co-efficient of determination) ', model_NYC_Food.score(input_NYC_OnlyFood,output_NYC_param_Onlyfood))
print('RMSE', np.sqrt(mean_squared_error(output_NYC_param_Onlyfood, predicted_output_NYC_OnlyFood_Model)))
print('Base RMSE', np.sqrt(mean_squared_error(output_NYC_param_Onlyfood, output_mean_NYC_param_OnlyFood)))
print('-------------------------End of Food and Price calculation-------------------------------')
print('----------------Dropping the service column for model calculation--------------------------')

input_NYC_droping_service_param = df_NYC[['Food','Decor','East']]
output_NYC__droping_service_param = df_NYC[['Price']]
model_NYC_Dropping_Service = LinearRegression()
model_NYC_Dropping_Service.fit(input_NYC_droping_service_param, output_NYC__droping_service_param)
#Evaluating
Predict_Unit_Price_dropping_service = model_NYC_Dropping_Service.predict(input_NYC_droping_service_param)
#print(Predict_Unit_Price_dropping_service)
mean_output_without_service = np.repeat(np.mean(output_NYC__droping_service_param), len(output_NYC__droping_service_param))
#print(mean_output_without_service)
#Validation and Performancecheck 
print('Coefficient', model_NYC_Dropping_Service.coef_, sep=':')
print('Intercept', model_NYC_Dropping_Service.intercept_, sep=':')
print('Score (RSquared)(Co-efficient of determination)',  model_NYC_Dropping_Service.score(input_NYC_droping_service_param,output_NYC__droping_service_param), sep=':')
print('RMSE', np.sqrt(mean_squared_error(output_NYC__droping_service_param,Predict_Unit_Price_dropping_service)), sep=':')
print('Base RMSE', np.sqrt(mean_squared_error(output_NYC__droping_service_param,mean_output_without_service)), sep=':')
```

```python
#Importing basic library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import math 

#importing linear regression related library 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#Loading the data

df_P301 = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\P301.txt', delimiter='\t')
#print(df_P301.head())
#print(df_P301.tail())
#print(df_P301.describe())
print(df_P301.info())


#Linear Regression for the data set 
input_P301 = pd.DataFrame(df_P301['Units'])
output_P301 = pd.DataFrame(df_P301['Minutes'])
#print(input_P301)
model_P301 = LinearRegression()
model_P301.fit(input_P301,output_P301)

#Prediction 
predicted_output_P301 = model_P301.predict(input_P301)

#Mean output value
output_mean_P301 = np.repeat(np.mean(df_P301['Minutes']), len(df_P301['Minutes']))
#print(output_mean_P301)

#Validation and Performance 
print('Coefficient', model_P301.coef_,sep=':')
print('Intercept', model_P301.intercept_, sep=':')
print('Score(RSquared)(Co-efficient of determination) ', model_P301.score(input_P301, output_P301), sep=':')
print('RMSE', np.sqrt(mean_squared_error(output_P301,predicted_output_P301)))
print('Base RMSE', np.sqrt(mean_squared_error(output_P301,output_mean_P301)))

#Testing for each individual computer repairing effort 

df_301_Test = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\P301 - Test.txt', delimiter='\t')
#print(df_301_Test.head())
#print(df_301_Test.tail())
#print(df_301_Test.describe())
#print(df_301_Test.info())
input_P301_Test = pd.DataFrame(df_301_Test['Units'])
#We have to test based on the created model named model_P301
Predicted_unit_price = model_P301.predict(input_P301_Test)
print(Predicted_unit_price)
print('Difference', np.round(np.subtract(Predicted_unit_price[1], Predicted_unit_price[0]),2), sep=':')
print('Difference', np.round(np.subtract(Predicted_unit_price[2], Predicted_unit_price[1]), 2), sep=':')

#Estimated time for repairing 4 computer 
df_301_Test_4Comp = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\P301 - Test - 4Comp.txt', delimiter='\t')
df_301_Test_4Comp
Predicted_Timeline = model_P301.predict(df_301_Test_4Comp)
print(Predicted_Timeline)
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


#Load the data
df_submission = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\submission.csv')
print(df_submission)
#print(df_submission.head())
#print(df_submission.tail())
#print(df_submission.describe())
#print(df_submission.info())

input_dataSet = df_submission[['PassengerId']]
output_dataSet = df_submission[['Survived']]

#print(input_dataSet)
#print(output_dataSet)

#Start decision tree
model_clf_Gini = DecisionTreeClassifier(criterion='gini', random_state=30, max_depth=3, max_leaf_nodes=5)
#model_clf_Entropy = DecisionTreeClassifier(criterion='entropy', random_state=30, max_depth=3, max_leaf_nodes=5)
model_clf_Gini.fit(input_dataSet, output_dataSet)
#model_clf_Entropy.fit(input_dataSet, output_dataSet)
#Prediction 
predicted_output = model_clf_Gini.predict(input_dataSet)
#print(predicted_output)

#Validation print
print('Confusion Matrix', confusion_matrix(output_dataSet, predicted_output))
print('Accuracy Score', accuracy_score(output_dataSet, predicted_output))
print('Classification Report', classification_report(output_dataSet, predicted_output))
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import LabelEncoder


#Loading the data
df_drug = pd.read_csv('E:\\Document\\Pre-RequisitePrgrame\\GITAA Training\\Notes\\Week 5\\DataSet\\drug200.csv')
#print(df_drug.describe())
#print(df_drug.info())
print(df_drug)

#Label encoding for categorical variable

lbl_en_Sex = LabelEncoder()
lbl_en_BP = LabelEncoder()
lbl_en_chol = LabelEncoder()
lbl_en_Drug = LabelEncoder()

df_drug['BP_New'] = lbl_en_BP.fit_transform(df_drug['BP'])
df_drug['Cholesterol_New'] = lbl_en_chol.fit_transform(df_drug['Cholesterol'])
df_drug['Drug_New'] = lbl_en_Drug.fit_transform(df_drug['Drug'])
df_drug['Sex_New'] = lbl_en_Sex.fit_transform(df_drug['Sex'])

df_drug.drop(['BP','Cholesterol','Drug','Sex'], axis=1, inplace=True)

copy_df_drug = df_drug.copy()

#Linear Regression 
output_data = df_drug.drop(['BP_New','Cholesterol_New','Sex_New','Age','Na_to_K'], axis=1)
input_data = df_drug.drop(['Drug_New'], axis=1)

print('data')
print(output_data)

model_Linear_Regression = LinearRegression()
model_Linear_Regression.fit(input_data, output_data)
print('Linear Regression')
print(model_Linear_Regression.score(input_data,output_data))
#print(model_Linear_Regression.predict(input_data))
print(model_Linear_Regression.coef_)
print(model_Linear_Regression.intercept_)
print(np.sqrt(mean_squared_error(output_data, model_Linear_Regression.predict(input_data))))

#Decision Tree

modelDecisionTree = DecisionTreeClassifier(criterion='gini', max_depth=4)
modelDecisionTree.fit(input_data,output_data)
predict_output = modelDecisionTree.predict(input_data)
print('Decision Tree')
print('Accuracy', accuracy_score(output_data,predict_output), sep=':')
print('Confusion Matrix', confusion_matrix(output_data, predict_output), sep=':')

#Random Forest 
classifier = RandomForestClassifier()
classifier.fit(input_data, output_data)
predicted_output_randomforest = classifier.predict(input_data)
print('Random Forest')
print('Accuracy score', accuracy_score(output_data, predicted_output_randomforest))

```
