```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data from an Excel file
measurements = pd.read_excel('Link 2 Measurements.xlsx')
print(measurements)
```

```python
# Separate the measurements for each tank
tank1_measurements = measurements['h1'].values
tank2_measurements = measurements['h2'].values
tank3_measurements = measurements['h3'].values
tank4_measurements = measurements['h4'].values
```

```python
measurements = measurements.to_numpy()
```

```python
h = measurements[0,:]
Ar = [28, 32, 28, 32]
ar = [0.071, 0.057, 0.071, 0.057]
g = 981

T = []
for i in range(len(h)):
    Time_const = (Ar[i] / ar[i]) * np.sqrt((2 * h[i]) / g)
    T.append(Time_const)

print(T)

A = np.array([[-1 / T[0], 0, Ar[2] / (Ar[0] * T[2]), 0],
              [0, -1 / T[1], 0, Ar[3] / (Ar[1] * T[3])],
              [0, 0, -1 / T[2], 0],
              [0, 0, 0, -1 / T[3]]])
print(A)
gamma_1 = 0.7
gamma_2 = 0.6
k1 = 3.33
k2 = 3.35

B = np.array([[(gamma_1 * k1) / Ar[0], 0],
              [0, (gamma_2 * k2) / Ar[1]],
              [0, ((1 - gamma_2) * k2) / Ar[2]],
              [((1 - gamma_1) * k1) / Ar[3], 0]])
print(B)
u = np.array([[3], [3]])

# Load measurements from Excel file
measurements = pd.read_excel('Link 2 Measurements.xlsx')
z_true = measurements[['h1', 'h2', 'h3', 'h4']].values


# Initialize variables
num_tanks = 4
num_measurements = len(z_true)
X_post = np.ones((num_tanks, 1))  # Initial state estimate [[12.4],[12.7],[1.8],[1.4]]
P_post = 1e5 * np.eye(num_tanks)  # Initial error covariance
Q = 8 * np.eye(num_tanks)  # Process noise covariance
R = 2 * np.eye(4)  # Measurement noise covariance
#R = 2 * np.eye(2)  # Measurement noise covariance


# Define the measurement matrix
# H = np.eye(num_tanks)
kc=0.5
H=[[kc,0,0,0],[0,kc,0,0],[0,0,kc,0],[0,0,0,kc]]
#H=[[kc,0,0,0],[0,kc,0,0]]
H=np.array(H)
# Initialize lists to store values for plotting
X_post_list,X_prior_list = [],[]
P_post_list = []
P_pri_list = []
Kalman_gain_list = []

# Kalman filter loop
for k in range(num_measurements):
    # Prediction step
    X_pri = np.dot(A, X_post) + np.dot(B, u[:].reshape(-1, 1))
    P_pri = np.dot(np.dot(A, P_post), A.T) + Q
    X_prior_list.append(X_pri.flatten())
    #print(X_pri)
    # Update step
    K = np.dot(np.dot(P_pri, H.T), np.linalg.inv(np.dot(np.dot(H, P_pri), H.T) + R))
    Z_est = np.dot(H, X_pri)
    E = z_true[k,:4] - Z_est.flatten()
    #E = z_true[k,:2] - Z_est.flatten()
    
    X_post = X_pri + np.dot(K, E.reshape(-1, 1))
    P_post = P_pri - np.dot(np.dot(K, H), P_pri)

    # Append values to the lists
    X_post_list.append(X_post.flatten())
    P_post_list.append(np.trace(P_post))
    P_pri_list.append(np.trace(P_pri))
    Kalman_gain_list.append(np.trace(K))

    # Stop condition: X_post - X_pri approx = 0
    if np.allclose(X_post, X_pri):
        break

# Final estimated states
estimated_states = X_post.flatten()
print("Estimated states:", estimated_states)
```

```python
def plot_filterwise_tank_data(subplot_num, time_steps, original_data, filter_results, tank_measurements, tank_name, marker):
    """
    Plots the tank data for a specific filter.

    Parameters:
        subplot_num (int): The subplot number for the current tank.
        time_steps (numpy.ndarray): Array of time steps.
        original_data (numpy.ndarray): Array of original tank data.
        filter_results (numpy.ndarray): Array of filter results for the current tank.
        tank_measurements (numpy.ndarray): Array of tank measurements for the current tank.
        tank_name (str): Name of the current tank.
        marker (str): Marking the tank wise plot

    Returns:
        None
    """
    # Create a subplot for the current tank
    plt.subplot(1, 4, subplot_num) 
    # Plot the original tank data
    plt.plot(time_steps, original_data, label='Original Data', color='blue')  
    # Plot the filter results
    plt.plot(time_steps, filter_results, label='Filter Results', color='red')  
    # Plot the tank measurements
    plt.scatter(time_steps, tank_measurements, label='Measurements', color='green', marker=marker, s = 30) 
    # Set the x-axis label
    plt.xlabel('Time Steps')  
    # Set the y-axis label
    plt.ylabel('Tank Level') 
    # Set the title for the subplot
    plt.title(tank_name)  
    # Show the legend
    plt.legend()  
```

```python
time_steps = np.arange(len(tank1_measurements))
```

```python
# Initialize the plot for original data and Kalman filter results
plt.figure(figsize=(12, 8))
plt.suptitle(f'Original and Kalman Filter Results')

plot_filterwise_tank_data(1, time_steps, tank1_measurements, np.array(X_post_list)[:, 0], tank1_measurements, 'Tank 1', '+')
plot_filterwise_tank_data(2, time_steps, tank2_measurements, np.array(X_post_list)[:, 1], tank2_measurements, 'Tank 2', 'o')
plot_filterwise_tank_data(3, time_steps, tank3_measurements, np.array(X_post_list)[:, 2], tank3_measurements, 'Tank 3', 's')
plot_filterwise_tank_data(4, time_steps, tank4_measurements, np.array(X_post_list)[:, 3], tank4_measurements, 'Tank 4', 'x')

# Adjust the subplot layout, so the plots do not overlap
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()   
```
