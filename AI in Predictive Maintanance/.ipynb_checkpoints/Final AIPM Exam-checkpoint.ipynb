```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()

# Create a DataFrame from the data
iris_df = pd.DataFrame(data.data, columns=data.feature_names)

# Add the target column to the DataFrame
iris_df['target'] = data.target

# Separate features and target variable
features = iris_df.drop('target', axis=1)
target = iris_df['target']

# Calculate feature contributions using correlation with target
contributions = features.corrwith(target)

# Sort the contributions in descending order
sorted_contributions = contributions.abs().sort_values(ascending=False)

# Create the contribution plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_contributions.index, sorted_contributions.values)
plt.xlabel('Features')
plt.ylabel('Contribution')
plt.title('Contribution Plot for Iris Dataset')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
