# ODE Solving

```python
import sympy as sp

def solve_ode(y, t, lhs, rhs, y0 = None, y_dash0 = None):
    Eq1 = sp.Eq( lhs, rhs)
    x = sp.symbols('x')


    print("ODE:")
    display(Eq1)


    # Solve the ODE symbolically
    solution = sp.dsolve(Eq1)

    # Print the general solution
    print("General solution:")
    print(solution)



    if y0 != None:
        
        # Print the particular solution with initial condition
        initial_conditions = {y.subs(t, 0): y0, sp.Derivative(y, t).subs(t, 0): y_dash0}  # Example initial conditions
        particular_solution = sp.dsolve(Eq1, ics=initial_conditions)
        # Print the particular solution
        print("\nParticular solution:")
        print(particular_solution)

        if (type(y0) is int or type(y0) is float) and (type(y_dash0) is int or type(y_dash0) is float):

            try:
                # Extract the particular solution
                particular_solution = particular_solution.rhs
                # Define the expressions for the derivative and second derivative
                dy_dt = sp.diff(particular_solution, t)
                d2y_dt2 = sp.diff(dy_dt, t)

                # Convert the sympy expressions to lambda functions
                y_func = sp.lambdify(t, particular_solution)
                dy_dt_func = sp.lambdify(t, dy_dt)
                d2y_dt2_func = sp.lambdify(t, d2y_dt2)

                # Define the time span for the plot
                t_values = np.linspace(0, 10, 100)

                # Evaluate the functions over the time span
                y_values = y_func(t_values)
                dy_dt_values = dy_dt_func(t_values)
                d2y_dt2_values = d2y_dt2_func(t_values)

                # Plot the functions
                plt.figure(figsize=(10, 6))
                plt.plot(t_values, y_values, label='y')
                plt.plot(t_values, dy_dt_values, label="dy/dt")
                plt.plot(t_values, d2y_dt2_values, label="d²y/dt²")
                plt.xlabel('t')
                plt.ylabel('Value')
                plt.title('Solution and Derivatives')
                plt.legend()
                plt.grid(True)
                plt.show()
            except NameError as e:
                print(f'Exception occuring while plotting: {str(e)}')
            except Exception as e:
                print(f'Exception occuring while plotting: {str(e)}')

    
# Define the symbol for the variable
t = sp.symbols('t')
# Define the symbol for the unknown function
y = sp.Function('y')(t)
```

# Bode Plot

```python
def plot_bode(sys, title):

    w, mag, phase = signal.bode(sys)

    fig, (mag_axis, phase_axis) = plt.subplots(1, 2, figsize=(18, 5))
    fig.suptitle(f'{title}', fontsize=22)
    mag_axis.set_title('Gain Plot', fontsize = 18)
    phase_axis.set_title('Phase Plot', fontsize = 18)
    
    mag_axis.semilogx(w, mag)  
    mag_axis.set_ylabel('Magnitude (dB)', fontsize=16)
    mag_axis.set_xlabel('Frequency (rad/sec)', fontsize=16)
    phase_axis.semilogx(w, phase)
    phase_axis.set_ylabel('Phase (deg)', fontsize=16)
    phase_axis.set_xlabel('Frequency (rad/sec)', fontsize=16)
    mag_axis.grid(True, which="both")
    phase_axis.grid(True, which="both")
    plt.tight_layout(pad = 2)
    plt.show()
    
z,p,k = (signal.tf2zpk([0, 1], [1, 30]))
sys = signal.ZerosPolesGain(z, p, 100)
plot_bode(sys, f'a) Real Poles C(s) = 100 / (30 + s)\nZeroes={list(z)} Poles={list(p)}')

z,p,k = (signal.tf2zpk([0, 1, 1], [1, 110, 1000]))
sys = signal.ZerosPolesGain(z, p, 100)
plot_bode(sys, f'b) Real Poles and Zeroes C(s) = 100 * (s + 1)/ (s\u00b2 + 110s + 1000)\nZeroes={list(z)} Poles={list(p)}')

z,p,k = (signal.tf2zpk([0,1, 10], [1, 3, 0]))
sys = signal.ZerosPolesGain(z, p, 10)
plot_bode(sys, f'c) Pole at the Origin C(s) = 10 * (10 + s) / (s\u00b2 + 3s)\nZeroes={list(z)} Poles={list(p)}')

z,p,k = (signal.tf2zpk([0,0,1, 0], [1, 12, 21, 10]))
sys = signal.ZerosPolesGain(z, p, -100)
p = np.round(p, 4)
plot_bode(sys, f'd) Repeated Real Poles, a negative constant C(s) = -100 * (s) / (s\u00b3 + 12s\u00b2 + 21s + 10)\nZeroes={list(z)} Poles={list(p)}')

z,p,k = (signal.tf2zpk([0, 1, 10], [1, 3, 50]))
sys = signal.ZerosPolesGain(z, p, 30)
p = np.round(p, 4)
plot_bode(sys, f'e) Complex Conjugate Poles C(s) = 30 * (s + 10) / (s\u00b2 + 3s + 50)\nZeroes={list(z)} Poles={list(p)}')

z,p,k = (signal.tf2zpk([0, 1, 1, 25], [1, 100, 0, 0]))
sys = signal.ZerosPolesGain(z, p, 4)
z = np.round(z, 4)
plot_bode(sys, f'e) Multiple Poles at Origin, Complex Conjugate Zeroes C(s) = 4 * (s\u00b2 + s + 25) / (s\u00b3 + 100s\u00b2)\nZeroes={list(z)} Poles={list(p)}')
```

# ODE to State Matrix

```python
#  10y(t) + 11dy(t)/dt + 6* d^2y(t)/dt^2 +  d^2y(t)/dt^2 = u(t)

import sympy as sp
# Define the symbols
t = sp.symbols('t')
x1, x2, x3 = sp.symbols('x1 x2 x3')
u = sp.Function('u')(t)

# Define the ODE
ode = sp.Eq(10 * x1 + 11 * sp.diff(x1, t) + 6 * sp.diff(x2, t) + sp.diff(x3, t, t), u)

# Formulate the state space equations
state_eqs = [sp.Eq(sp.diff(x1, t), x2),
             sp.Eq(sp.diff(x2, t), x3),
             sp.Eq(sp.diff(x3, t, 3), -10 * x1 - 11 * x2 - 6 * x3 - 8 * u)]

# Extract the system matrices
variables = [x1, x2, x3]
derivatives = [sp.diff(var, t, 1) for var in variables]
A, B = sp.linear_eq_to_matrix(state_eqs, variables)
# C = sp.zeros(1, len(variables))
# C[0] = 1
# D = sp.zeros(C.shape[0], B.shape[1])
C, D = sp.linear_eq_to_matrix([x1], variables)

# Print the state space matrices
print("A matrix:")
print(A)
print("\nB matrix:")
print(B)
print("\nC matrix:")
print(C)
print("\nD matrix:")
print(D)
```

# ODE to Transfer Function

```python
import sympy as sp
from sympy.abc import s

# Define the symbols
Y = sp.Function('Y')(s)
U = sp.Function('U')(s)

# Define the Laplace transforms of the derivatives
Y_prime = s * Y
Y_double_prime = s**2 * Y

# Define the ODE in the Laplace domain
ode = 10 * Y + 11 * Y_prime + 6 * Y_double_prime + s**3 * Y - U

# Solve the ODE for the transfer function
transfer_function = sp.solve(ode, Y)[0] / U

# Print the transfer function
print("Transfer Function:")
print(transfer_function)
```
