#### Import the packages.

```python
import pandas as pd
from kafka import KafkaProducer
import json
import time
import random
from pandas import Timestamp
from datetime import datetime
```

```python
yelp_reviews = pd.read_json('/content/sample_data/small_yelp_reviews.json', lines = True)
data = yelp_reviews.to_dict("reviews")
```

```python
# Define the Kafka topic
topic = "yelp_reviews"
# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers="localhost:9092")
print('Producer sending data:')
producer.flush()
# Iterate over the data and send each row to Kafka
for idx, data_row in enumerate(data, start=1):
    print(data_row)
    # Format the 'date' field to a specific datetime format
    data_row['date'] = data_row['date'].strftime('%Y-%m-%d %H:%M:%S')
    # Convert the row to a JSON-encoded message
    message = json.dumps(data_row).encode("utf-8")
    # Send the message to the Kafka topic
    producer.send(topic, message)
    # Introduce a 10-second delay every 10 records
    if idx % 10 == 0:
        time.sleep(10)
# Flush the producer
producer.flush()
```
