```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# set the style for plotting
sns.set_theme(
    context="notebook",
    style="white",
    palette="colorblind",
    rc={"figure.figsize": (14, 10), "figure.dpi": 300},
)

sns.set_palette(  # creating a palette of colors
    [
        "#000000",
        "#E69F00",
        "#56B4E9",
        "#009E73",
        "#F0E442",
        "#0072B2",
        "#D55E00",
        "#CC79A7",
    ]
)
```

```python
df = pd.read_csv("soln_data.csv")  # reading the data
df.head()
```

# Importing the necessary scikit-learn libraries

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Creating variables

X = df.drop(["y1", "y2", 'y3'], axis=1)
y = df[["y1"]]

X_std = StandardScaler().fit_transform(X.values)

# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2, random_state=42)
X_train.shape
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

y_pred = linear_model.predict(X_test)

print("Mean squared error: {}".format(mean_squared_error(y_test, y_pred)))
print("R2 score: {}".format(r2_score(y_test, y_pred)))
```

```python
alpha = 0.1 # setting the value of alpha

ridge_model = Ridge(alpha=alpha) # creating the model
ridge_model.fit(X_train, y_train) # fitting the model

y_pred = ridge_model.predict(X_test) # predicting the values

print("R2 score: {}".format(r2_score(y_test, y_pred))) # printing the R2 score
print("MSE: {}".format(mean_squared_error(y_test, y_pred))) # printing the MSE
```

```python
alpha = 0.1 # setting the value of alpha

lasso_model = Lasso(alpha=alpha) # creating the model
lasso_model.fit(X_train, y_train) # fitting the model

y_pred = lasso_model.predict(X_test) # predicting the values

print("R2 score: {}".format(r2_score(y_test, y_pred))) # printing the R2 score
print("MSE: {}".format(mean_squared_error(y_test, y_pred))) # printing the MSE
```

# Comaprison

By comparing the Ridge and Lasso models, it is evident that for this particular dataset, the performance is nearly equivalent with Ridge performing very marginally better than Lasso. However, Lasso is slightly more interpretable as it can be used to select the most important features. In this case, the most important features are x1, x2, x3. The other features are not important and can be removed from the model.

# Understanding the effect of $\alpha$

```python
# to test the effect of alpha on the model and to plot it 

coefficients = []

for i in range(1,11):
    alpha = i/10 # setting the value of alpha
    lasso_model = Lasso(alpha=alpha) # creating the model
    lasso_model.fit(X_train, y_train) # fitting the model
    coefficients.append(lasso_model.coef_) # appending the coefficients
    
coefficients = np.array(coefficients) # converting the list to array
coefficients = coefficients.reshape(-1,1)
```

```python
coefficients_df = pd.DataFrame(coefficients, columns=['coefficients'])                                                          # creating a dataframe of the coefficients
coefficients_df['alpha'] = np.asarray([[j/10 for i in range(6)] for j in range(1,11)]).reshape(-1,1)                            # adding the alpha values to the dataframe
coefficients_df['feature'] = np.asarray([['x1', 'x2', 'x3', 'x4', 'x5', 'x6'] for j in np.arange(0.1, 1.1, 0.1)]).reshape(-1,1) # adding the feature names to the dataframe
coefficients_df.head(6)
```

```python
sns.barplot(data=coefficients_df, x="alpha", y="coefficients", hue="feature") # plotting the coefficients
plt.title("Effect of alpha on coefficients - LASSO")
```

# Effect of $\alpha$ 

We can observe that, as $\alpha$ increases, the coefficients of the model decrease. This is because the model is penalized more and more for having large coefficients. A similar analysis can be done for Ridge.
