<b> Supporting Function Calls</b>

```python
from sklearn.metrics import rand_score, silhouette_score
from sklearn.metrics import accuracy_score
```

```python
def evalCluster(X, y, y_pred, index):
    '''
    Scoring a clustering
      X: input feature
      y: true label
      y_pred: predicted label
      prints Rand Score, Silhouette Score and Accuracy
    '''
    print('For : ', index)
    print("Rand_score")
    print("Blobs: {:.4f}".format(rand_score(y, y_pred)))
    print("\n")
    print("Silhouette_score")
    print("Blobs: {:.4f}".format(silhouette_score(X, y_pred)))
    print('\n')
    print(f'Accuracy : {accuracy_score(y,y_pred):.4f}')
```

### Q1) Apply the k-means clustering algorithm to the data and evaluate its performance. Use the class column as the true class.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state = seed,n_init='auto'
    
)

m1 = kmeans.fit(X, y)
m1_pred = m1.predict(X)
```

```python
np.unique(m1_pred)
```

```python
# np.mean(y == m1_pred)
```

```python
evalCluster(X, y, m1_pred, 'K-Mean')
```

### Q2 Use hierarchical clustering on the data and report the effect of the parameters linkage and distance metric on the number of clusters obtained.

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
clust = AgglomerativeClustering(
    n_clusters=5,               # Needs to be None if distance_threshold is used
    metric= 'euclidean',
    linkage = 'single',
    # distance_threshold = 1250
)

m2_pred = clust.fit_predict(X,y)
```

```python
evalCluster(X, y, m2_pred,'Hierarchical')
```

Linkage and Dendogram

```python
# %%time
Z = linkage(
    y = X,       # Input
    method='average', # Linkage
    metric='euclidean' 
)

print(Z)

fig, ax = plt.subplots()
R = dendrogram(Z, ax=ax);
ax.set_title("Dendrgoram of the Clustering - Blobs", size=14)
```

From Dendogram, we can observe 2 things. Number of possible group with different colors and the threshold in y axis to get those seperations. 

```python
clust = AgglomerativeClustering(
    n_clusters=None,               # Needs to be None if distance_threshold is used
    metric= 'euclidean',
    linkage = 'single',
    distance_threshold = 1200
)

m2_pred = clust.fit_predict(X,y)
```

```python
np.unique(m2_pred)
```

```python
evalCluster(X, y, m2_pred, 'Hierarchical')
```

### Q3 Cluster the given data using spectral clustering. Are the obtained clusters better than the ones obtined from K-means or hierarchical clustering?

```python
from sklearn.cluster import SpectralClustering
```

```python
sclust = SpectralClustering(
    n_clusters = 5,
    affinity='nearest_neighbors', # Method to create graph
    random_state = 42,
    n_neighbors=50
)

m3_pred = sclust.fit_predict(X,y)
```

```python
np.unique(m3_pred)
```

```python
evalCluster(X, y, m3_pred, 'Spectral')
```

### Q4 Visualize the clusters using the UMAP or tSNE algorithm.

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

<b>Using tSNE Method</b>

```python
X_scaled = StandardScaler().fit_transform(X)

tsne = TSNE(
        n_components=2,
        perplexity = 40,
        n_iter=500,
        n_iter_without_progress=150,
        n_jobs=2,
        random_state=seed,
    )

tsne_transform = tsne.fit_transform(X_scaled, y)
```

```python
df_tsne = pd.DataFrame(data = tsne_transform, columns=[f'Component_{i}' for i in range(2)])
df_tsne['y'] = y
```

```python
cmap = sns.color_palette('hls', n_colors=5)
fig, ax = plt.subplots(figsize=(7,7))
sns.scatterplot(
    data=df_tsne,
    x='Component_0',
    y='Component_1',
    hue='y',
    palette=cmap
)
plt.title('tSNE on the given dataset', size=18)
plt.show()
```

<b>using UMAP</b>

```python
from umap import UMAP
# import umap.plot
```

```python

X_scaled = StandardScaler().fit_transform(X)
```

```python
n_neighbors = 60
min_dist = 0.1

reducer = UMAP(
    n_neighbors=n_neighbors,
    min_dist=min_dist,
    random_state=seed
)

reducer.fit(X_scaled)
```

```python
umap_transform = reducer.transform(X_scaled)

df_umap = pd.DataFrame(data = umap_transform, columns=[f'Component_{i}' for i in range(2)])
df_umap['y'] = y
```

```python
cmap = sns.color_palette('hls', n_colors=5)
fig, ax = plt.subplots(figsize=(7,7))
sns.scatterplot(
    data=df_umap,
    x='Component_0',
    y='Component_1',
    hue='y',
    palette=cmap
)
plt.title('UMAP on the given dataset', size=18)
plt.show()
```
