```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import rand_score, accuracy_score, silhouette_score
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram
from mpl_toolkits.mplot3d import Axes3D


sns.set(style='darkgrid')
```

# Loading the data and splitting into Input and Output dataframes

```python
dat = pd.read_csv('18.csv')
dat.head()
```

```python
X = dat.iloc[:,:-1]
Y = dat.iloc[:,-1]

X_scaled = StandardScaler().fit_transform(X)
```

```python
seed = 50
```

```python
np.unique(dat['y'])
```

# Q1-> KMeans clustering

```python
kmeans = KMeans(
    n_clusters=3,
    random_state = seed
)

kmeans_pred = kmeans.fit_predict(X_scaled,Y)

plt.figure(figsize=(6,6))

sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], c=kmeans_pred, alpha=0.6)
plt.legend(labels=['Predicted'])
```

```python
# Scoring

print("Accuracy_score")
print(np.round(accuracy_score(Y,kmeans_pred), 4))

print("Rand_score")
print(round(rand_score(Y, kmeans_pred), 4))

print("Silhouette_score")
print(round(silhouette_score(X_scaled, kmeans_pred), 4))
```

# Q1-> conclusions:

1. Apart from the Silhouette Score the Rand and Accuracy scores are better.
2. Silhouette score is closer to 0 so there is some amount of overlapping.
3. Rand_score suggests the level of clustering achieved is only just above average.

# Q2-> Hierarchical Clustering

```python
def hc_ac(metric, linkage, X_scaled, Y):
    cluster = AgglomerativeClustering(
        n_clusters=3,               # Needs to be None if distance_threshold is used
        metric = metric,
        linkage = linkage,
    )

    hc_pred = cluster.fit_predict(X_scaled,Y)
    
    return hc_pred
```

```python
# Scoring
def scoring_hc_ac(Y, hc_pred, X_scaled):
    print("Accuracy_score")
    print(round(accuracy_score(Y, hc_pred), 4))

    print("Rand_score")
    print(round(rand_score(Y, hc_pred),4))

    print("Silhouette_score")
    print(round(silhouette_score(X_scaled, hc_pred),4))
```

```python
def get_clusters_dendrogram(metr, link, X_scaled):
    l = linkage(y = X_scaled, method=link, metric=metr)


    fig, ax = plt.subplots(figsize=(10,12))
    d = dendrogram(l, ax=ax);

    ax.set_title("Dendrogram of the clustering is below", size=14)
    plt.show()
```

# combinatorical searching on agglomerative clustering

```python
linkg = ['complete', 'average', 'single', 'ward']
metrc = ['euclidean', 'manhattan', 'cosine', 'l1', 'l2', 'precomputed']
count = 0
hc_pred_arr = np.zeros((656,24))
for link in linkg:
    for metr in metrc:
        print('Combination #',count+1)
        count+=1
        try:        
            hc_pred2 = hc_ac(metr, link, X_scaled, Y)
            print('"Linkage"-',link,'and "metric"-',metr)
            print('------------------------------------')
            scoring_hc_ac(Y, hc_pred2, X_scaled)
            print('------------------------------------')
            print('\n')
            hc_pred_arr[:,count-1] = hc_pred2.reshape(656,1)
        except ValueError:
            print('Incompatible metric and linkage combination provided: ', metr,' with ',link)
            print('\n')
        except:
            print('Please check the module parameters')
```

# No. of clusters per dendrogram per linkage and metric combination

```python
linkg = ['complete', 'average', 'single', 'ward']
metrc = ['euclidean', 'manhattan', 'cosine', 'l1', 'l2', 'precomputed']
count = 0
for link in linkg:
    for metr in metrc:
        print('Combination #',count+1)
        count+=1
        try: 
            print('"Linkage"-',link,'and "metric"-',metr)
            get_clusters_dendrogram(metr, link, X_scaled)
            print('\n')
            
        except ValueError:
            print('Incompatible metric and linkage combination provided: ', metr,' with ',link)
            print('\n')
        except:
            print('Please check the module parameters')
```

# Q2-> Conclusion: Effects of different linkage and metric combination on the no. of clusters from the dendrogram are mentioned below-

1. Combination #19: ward and euclidean based agglomerative clustering has produced the best/distinct 3 clusters as expected just about y-axis=30.
2. Followed by combination #3: complete and cosine based agglomerative clustering also has produced 3 clusters as expected near y-axis=1.75.
3. Combination #1: complete and euclidean based agglomerative clustering has shown 2 clusters or just about 3 clusters closer to y-axis=14.
4. Combination #9: average and cosine based agglomerative clustering shows 4 clusters or 2 clusters closer to y-axis=1.0.
5. Combinations #7[average and euclidean], 13[single and euclidean], 15[single and cosine]: produced dendrograms that doesn't show even 1 cluster properly.
6. Other 17 linkage and metric combinations are not compatible hence no linkage and no dendrogram.

# Q3-> Spectral Clustering

```python
from sklearn.cluster import SpectralClustering

spectral_cluster = SpectralClustering(
    n_clusters = 3,
    affinity='nearest_neighbors', # Method to create graph
    random_state = 30,
    n_neighbors=10
)

y_pred_sc = spectral_cluster.fit_predict(X_scaled, Y)

sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=y_pred_sc)
plt.legend(title='Predicted')
```

```python
# Spectral Scoring

print("Accuracy_score")
print(np.round(accuracy_score(Y, y_pred_sc), 4))

print("Rand_score")
print(round(rand_score(Y, y_pred_sc), 4))

print("Silhouette_score")
print(round(silhouette_score(X_scaled, y_pred_sc), 4))
```

# Q3 conclusion:

KMeans shows better performance than Spectral and Hierarchical Clustering algorithms in terms of Accuracy, Rand and Silhouette scores
